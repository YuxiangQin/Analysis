```python
def get_data(file_path):
    '''Return a dataframe from .sql file stored in @file_path'''
    creds, project = google.auth.default()
    client = Client(project,creds)

    with open(file_path, "r") as file:
        query = file.read()

    df = client.query(query).to_dataframe()
    return df
```

```python
def join_strings(row):
    '''Join strings with comma, skip null values'''
    non_null_values = [str(value) for value in row if pd.notnull(value)]
    return ', '.join(non_null_values)

def clean_emails(text):
    '''Clean leading and trailing spaces then join together with comma'''
    if pd.isnull(text):
        return None
    else:
        return ', '.join([w.strip() for w in text.split(',')])

def clean_data(df):
    df['Chargebacks'] = df[['Late_T', 'Reject_T', 'TCB_T']].apply(join_strings, axis=1)
    df['VendorEmails'] = df['VendorEmails'].apply(clean_emails)
    df['OwnerEmail'] = df['OwnerEmail'].apply(clean_emails)
    return df
```

```python
def html_template(file_path, mapping):
    '''Returns a string of html file from @file_path with its content replaced using @mapping'''
    with open(file_path, 'r') as file:
        template = Template(file.read())

    return template.safe_substitute(mapping)
```

```python
def create_email(sender, to_str, cc_str, html_body, 
                 subject='Action Required: Drop Ship Vendor Performance Alert',
                 attachment=None):
    msg = MIMEMultipart()
    msg['From'] = sender
    msg['To'] = to_str
    if cc_str is not None:
        msg['Cc'] = cc_str   
    msg['Subject'] = subject
    msg.attach(MIMEText(html_body, 'html'))
    if attachment is not None:
        msg.attach(attachment)

    return msg

def send_email(msg):
    with smtplib.SMTP('smtp.wal-mart.com') as server:
        server.send_message(msg)
```

```python
def export_excel(df_dict):
    '''Convert DF to Excel (Used by emailing UDFs)'''
    with io.BytesIO() as buffer:
        with pd.ExcelWriter(buffer) as writer:
            for sheetname, data in df_dict.items():
                data.to_excel(writer, sheetname, index=False)
        return buffer.getvalue()
```

```python
# send log
def send_log(df_input, log_config):
    df = df_input[log_config['cols']]
    year_month = f"{df.iloc[0]['year']}-{df.iloc[0]['month']}"
    email_date = datetime.now(pytz.timezone('America/Los_Angeles')).strftime("%Y-%m-%d")
    run_stats = df['Comments'].fillna(" ").value_counts().reset_index().rename(columns={'index':'Status','Comments':'Vendor'}).to_html(index=False)
    log_mapping = {
        "year_month": year_month,
        "email_date": email_date,
        "run_stats": run_stats
    }
    html_body = html_template(log_config['html_template'], log_mapping)
    logname = log_config['filename']
    filename = f"{year_month}_{logname}_send_on_{email_date}.xlsx"
    attachment = MIMEApplication(export_excel({'Summary': df}))
    attachment['Content-Disposition'] = f'attachment; filename={filename}'
    # save log file
    df.to_excel(log_config['filepath'] + '\\' + filename, index=False)
    # send log email
    msg = create_email(log_config['sender'], log_config['to'], log_config['cc'], 
                       html_body, log_config['subject'], attachment)
    send_email(msg)
```

```python
#UDFs

#1. Get data from BQ
def get_data(query):
    result = dataset.load(name='Google_BigQuery_Connector',query=query)
    return result

#2. Generate Vendor-Recipients Dictionary
def get_recipients(metrics_data):
    def lol(a,b):
        if not b:
            b = ''
        
        l = [s.strip() for s in a.split(',')]
        b = b.strip()
        new_list = [l] + [[b]]
        return new_list
    emaildf = metrics_data[metrics_data['Comments'].isna()][['OwnerEmail','VendorEmails']].copy()
    emaildf['new'] = emaildf.apply(lambda x: lol(x.VendorEmails,x.OwnerEmail), axis=1)
    emaildict = emaildf['new'].to_dict()    # dictionary of : {VendorID: [[vendoremails],[owneremails]]}
    return emaildict

#3. Read HTML file
def get_html_template(emaildate,reportweek):
    with open(r"/data/DSV/Email_Templates/Lates_Email_Template.html") as file:
        template=file.read()
    return template

#4. Generate inputs for email
def per_vendor_email_params(vendorid, po_data_per_vendor, emaildate, metrics_df_per_vendor, template,emaildict):
   
    attach_dict = {"PO Data": po_data_per_vendor} 
    filename = str(vendorid)+' '+str(emaildate)+" Late(or)Unshipped PO Details.xlsx"    
    email_sub = "WMT DSV Late/Unshipped Order Notification: " + str(vendorid) + ' - '+ emaildate
    #2.2) Replacing default HTML tags with custom attributes
    email_body = template
    metrics_tag_replace={'[Vendor_Name]':str(metrics_df_per_vendor['VENDOR_NAME'].item()),'[PO_Count]':str(metrics_df_per_vendor['PO_Count'].item())}
    for oldtag,newtag in metrics_tag_replace.items():
        email_body = email_body.replace(oldtag,newtag)
    to_list = emaildict[vendorid][0]
    cc_list = emaildict[vendorid][1]
    return (email_sub,to_list,cc_list,email_body,attach_dict,filename)

#5. Convert DF to Excel (Used by emailing UDFs)
def export_excel(df_dict):
  with io.BytesIO() as buffer:
    with pd.ExcelWriter(buffer) as writer:
        for sheetname,data in df_dict.items():
            data.to_excel(writer, sheetname, index=False)
    return buffer.getvalue()
```
