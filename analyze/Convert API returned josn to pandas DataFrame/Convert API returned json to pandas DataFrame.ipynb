```python
import pandas as pd
import json
```

```python
with open('data/QB_Response.json') as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
data['metadata']
```

```python
data['data']
```

```python
data['fields']
```

```python
# create dataframe use data['data']
df = pd.json_normalize(data['data'])
```

```python
df.head()
```

```python
# create id - column_name mapping use data['fields']
mapping = {field['id']: field['label'] for field in data['fields']}
```

```python
# rename column names
df.columns = [mapping[int(col.split('.')[0])] for col in df.columns]
```

```python
df.head()
```

```python
df.to_csv('data\output_file.csv', index=False)
```
